```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', None)

path = 'C:/VSC/JJaemni/CSV/'
```

```python
df = pd.read_csv(path + 'wanted_jobplanet.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

- `Preference`에서 Null값은 `Requirement` 데이터로 대체
- `StackTool`에서 Null값은 행 제거

```python
# Requirement tolist
req_list = df.Requirement.tolist()

# Preference tolist
pre_list = df.Preference.tolist()

# StackTool tolist
sta_list = df.StackTool.tolist()
```

```python
# Requirement
for idx, data in enumerate(req_list):
    print(f'{idx}. {data}')
    print('---------------------------')
```

```python
df.columns
```

```python
df2 = df.copy()
```

```python
# 17, 57, 98, 170행은 Null값으로 변경

index = [17, 57, 98, 170]
for idx in index:
    df2.loc[idx].iloc[3] = np.nan
    print(df2.loc[idx].iloc[3], end=' ')
```

```python
# Preference
for idx, data in enumerate(pre_list):
    print(f'{idx}. {data}')
    print('---------------------------')
```

```python
# 17, 42, 57, 65, 98, 127, 133, 197행은 Null값으로 변경

index = [17, 42, 57, 65, 98, 127, 133, 197]
for idx in index:
    df2.loc[idx].iloc[4] = np.nan
    print(df2.loc[idx].iloc[4], end=' ')
```

```python
# StackTool
for idx, data in enumerate(sta_list):
    print(f'{idx}. {data}')
    print('---------------------------')
```

```python
def Null(col):
    a = df2[df2[f'{col}'].isnull()].index
    print(f'\n# {col}\n{a}\n{len(a)}')

cols = ['Requirement', 'Preference', 'StackTool']
for col in cols:
    Null(col)
```
