- `Preference`에서 Null값은 `Requirement` 데이터로 대체
- `StackTool`에서 Null값은 행 제거

### Preference에서 Null값은 Requirement 데이터로 대체

```python
df2 = df.copy()
```

```python
df2.Preference[133]
```

```python
df2.Preference[133] = np.NaN
```

```python
idxx = df2[df2.Preference.isnull()].index
for idx in idxx:
    print(df2.loc[idx].Link)
```

```python
idxx = df2[df2.Preference.isnull()].index
for idx in idxx:
    df2.Preference[idx] = df2.Requirement[idx]
    print(idx)
```

### StackTool에서 Null값은 행 제거

```python
idxx = df2[df2.StackTool.isnull()].index
for idx in idxx:
    print(idx, end=' ')
```

```python
df2.dropna(axis=0, inplace=True)
df2.isnull().sum()
```

### Requirement에서 \n값만 있는 170행은 Preference값으로 대체
